```python
def print_stadistic(data):
    print("N Images: " + str(len(data["age"])))
    print("Mean: "+ str(data["age"].mean()))
    print("Std: "+ str(data["age"].std()))
    print("Min Age: "+ str(data["age"].min()))
    print("Max Age: "+ str(data["age"].max()))
    print("(%) female: " + str(data["gender"].value_counts()["F"]/len(data["gender"])))


```

```python

def balance_gender(data, min_images=59):
    male = data[data["gender"]=="M"]
    female = data[data["gender"]=="F"]
    
    male = male.iloc[0:min_images,:]
    female = female.iloc[0:min_images,:]

    data_balanced = pd.concat([male,female])

    return data_balanced

Y_ID1000 = pd.read_csv("/home/nnieto/Nico/Harmonization/data/final_data_split/Y_ID1000.csv")                # noqa
Y_eNKI = pd.read_csv("/home/nnieto/Nico/Harmonization/data/final_data_split/Y_eNKI_crop.csv")               # noqa
Y_Camcan = pd.read_csv("/home/nnieto/Nico/Harmonization/data/final_data_split/Y_CamCAN_crop.csv")           # noqa
Y_1000brains = pd.read_csv("/home/nnieto/Nico/Harmonization/data/final_data_split/Y_1000Gehirne_crop.csv")  # noqa




Y_1000brains_age_unbalanced_sex_balanced = balance_gender(Y_1000brains)
Y_Camcan_age_unbalanced_sex_balanced = balance_gender(Y_Camcan)
Y_ID1000_age_unbalanced_sex_balanced = balance_gender(Y_ID1000)
Y_eNKI_age_unbalanced_sex_balanced = balance_gender(Y_eNKI)

Y = pd.concat([Y_ID1000_age_unbalanced_sex_balanced,
                    Y_eNKI_age_unbalanced_sex_balanced,
                  Y_Camcan_age_unbalanced_sex_balanced,
                  Y_1000brains_age_unbalanced_sex_balanced
                  ])



sbn.swarmplot(data=Y, x="age", y="site", hue="gender", size=4)
```

```python
X_ID1000 = pd.read_csv("/home/nnieto/Nico/Harmonization/data/final_data_split/X_ID1000.csv")                # noqa
X_eNKI = pd.read_csv("/home/nnieto/Nico/Harmonization/data/final_data_split/X_eNKI_crop.csv")               # noqa
X_Camcan = pd.read_csv("/home/nnieto/Nico/Harmonization/data/final_data_split/X_CamCAN_crop.csv")           # noqa
X_1000brains = pd.read_csv("/home/nnieto/Nico/Harmonization/data/final_data_split/X_1000Gehirne_crop.csv")  # noqa


def retain_images(X,Y):
    return X.loc[Y.index, :]

X = pd.concat([retain_images(X_ID1000,Y_ID1000_age_unbalanced_sex_balanced),
                    retain_images(X_eNKI,Y_eNKI_age_unbalanced_sex_balanced),
                  retain_images(X_Camcan,Y_Camcan_age_unbalanced_sex_balanced),
                  retain_images(X_1000brains,Y_1000brains_age_unbalanced_sex_balanced)
                  ])



X.dropna(axis=1, inplace=True)
X.shape
```
